```python
import numpy as np
import mne
%matplotlib inline
import matplotlib.pyplot as plt
import scipy
from scipy.stats import hmean,trim_mean
```

```python
filename = "suj11_l5nap_day1.vhdr"
raw = mne.io.read_raw_brainvision(filename,preload=True)
raw.info
```

```python
"""To visualize the data"""
"""Unit of raw data is in voltage, thus scaling for EEG is 20*10^-6"""
scalingDict = dict(mag=1e-12, grad=4e-11, eeg=20e-6, eog=150e-6, ecg=5e-4,
                  emg=1e-3,ref_meg=1e-12,misc=1e-3,stim=1,
                  resp=1, chpi=1e-4)
raw.plot(start=0, #initale time to show
         duration=10.0, #time window(seconds to show),
         n_channels=20, #number of channels to plot at once
         scalings=scalingDict)
```

```python
"""To plot the power spectral density across data"""
raw.plot_psd(tmin=0.0, #inital time to show
             tmax=60.0, #end time to show
             fmin=0.0, #inital frequency to show
             fmax=200.0,
             area_mode='std' #change to 'range'
            )
```
