### Number of Epochs for all Subjects, all individual Runs
This is useful for keeping track of which files may be too small for certain classification algorithms.  
For example, Running KNN where number of neighbors = 10, but the number of epochs = 4. 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 

import os.path as op
from os import listdir

import warnings
warnings.simplefilter("ignore", DeprecationWarning)

#import mne to read EEG data 
import mne 
from mne.decoding import Vectorizer
```

```python
data_dir = "/Volumes/SB/eeg_data"
all_runs = [f for f in listdir(data_dir) if op.isfile(op.join(data_dir, f))]

all_epochs = {}
for i in all_runs:
    epo = mne.epochs.read_epochs(op.join(data_dir, i), preload=True);
    all_epochs[i] = len(epo.events)
```
